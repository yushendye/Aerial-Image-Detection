## Step 4: Now it's time to train!!
#### Be careful how you use the commands!!
#### --img: specify image size, in my project I tried 3 different image sizes. 416, 832 and 608
#### --batch is just batch size
#### --epochs are just the number of epochs 
#### --data is lication of coco.yaml file where you have configured your dataset in step 3
#### --cfg is the config file you are using for training. For this project I have used yolov4-csp.yaml file
#### --weights is the weight file which was saved during your training. Initially you can keep exclude this parameter.
#### --name is simply the name that will be given to directory where your weights would be saved
#### --logdir is directroy where you are going to save the model weights

### --hyp is hyperprameter file where you have explicitly mentioned the hyperparameters

```python
%cd /content/ScaledYOLOv4/
!python train.py --img 608 --batch 4 --epochs 700 --data /content/ScaledYOLOv4/data/coco.yaml --cfg /content/ScaledYOLOv4/models/yolov4-csp.yaml --weights '/content/drive/MyDrive/8_Nov_CSP_Scratch_608_3/exp0_yolov4-csp-results/weights/last.pt' --name yolov4-csp-results --logdir '/content/drive/MyDrive/9_Nov_CSP_Scratch_608/' --hyp /content/ScaledYOLOv4/data/hyp.scratch.yaml
```

## Step 5: Inference the model on your images!!
#### --conf is confidence you want your model to have about certain object being in the image. It works as a threshold
#### --source is the directory where your images are
